```python

import numpy as np
import matplotlib.pyplot as plt

# Fitness function
def fitness(x):
    return x * np.sin(x)

# Parameters
pop_size = 20
gens = 50
cross_rate = 0.8
mut_rate = 0.1
x_min, x_max = 0, 10

# Initial population
pop = np.random.uniform(x_min, x_max, pop_size)
best_history = []

for gen in range(gens):
    # Evaluate fitness
    fit = fitness(pop)

    # Select top 50%
    top = pop[np.argsort(fit)[-pop_size//2:]]

    # Create new population
    new_pop = []
    for _ in range(pop_size):
        if np.random.rand() < cross_rate:
            p1, p2 = np.random.choice(top, 2, replace=False)
            child = (p1 + p2) / 2
        else:
            child = np.random.choice(top)

        # Mutation
        if np.random.rand() < mut_rate:
            child += np.random.uniform(-1, 1)
        child = np.clip(child, x_min, x_max)
        new_pop.append(child)

    pop = np.array(new_pop)
    best_fit = np.max(fitness(pop))
    best_history.append(best_fit)
    print(f"Gen {gen+1}: Best Fitness = {best_fit:.4f}")

# Final best result
best_x = pop[np.argmax(fitness(pop))]
print(f"✅ Best x = {best_x:.4f}, f(x) = {fitness(best_x):.4f}")

# Plot
plt.plot(best_history)
plt.title("Fitness Over Generations")
plt.xlabel("Generation")
plt.ylabel("Best Fitness")
plt.grid(True)
plt.show()

```
